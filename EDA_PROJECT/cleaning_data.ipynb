```python
import re
```

```python
# Function to extract the first integer from a string
def extract_warranty_num(x):
    if pd.isna(x):
        return None
    s = str(x)
    m = re.search(r'(\d+)', s)   
    if m:
        return int(m.group(1))
    else:
        return None


df['Warranty_year'] = df['warranty'].apply(extract_warranty_num)


df_clean = df[df['Warranty_year'].isin([1,2,3])]


warranty_nums = df['Warranty_year'].tolist()

print(warranty_nums)

```

```python
df["discounted_price"] = df["discounted_price"].str.replace("₹", "", regex=False) \
                         .str.replace(",", "", regex=False) \
                         .str.strip()
df["discounted_price"] = pd.to_numeric(df["discounted_price"], errors="coerce")
```

```python
df["original_price"] = df["original_price"].str.replace("₹", "", regex=False) \
                         .str.replace(",", "", regex=False) \
                         .str.strip()
df["original_price"] = pd.to_numeric(df["original_price"], errors="coerce")
```

```python
df["number_of_reviews"] = df["number_of_reviews"].str.replace(",", "", regex=False)
df["number_of_reviews"] = pd.to_numeric(df["number_of_reviews"], errors="coerce").astype("Int64")
```

```python
df = df.drop("Warranty_numeric", axis=1)
```

```python
df["size_cm"] = df["screen_size"].str.extract(r"(\d+)\s*cm", expand=False)
df["size_inch"] = df["screen_size"].str.extract(r"\((\d+)\s*inch\)", expand=False)


df["size_cm"] = pd.to_numeric(df["size_cm"], errors="coerce").astype("Int64")
df["size_inch"] = pd.to_numeric(df["size_inch"], errors="coerce").astype("Int64")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
avg_reviews = df.groupby("brand")["number_of_reviews"].mean().sort_values(ascending=False)


plt.figure(figsize=(20, 10))  
sns.barplot(x=avg_reviews.index, y=avg_reviews.values, palette="viridis")
plt.title("Average Number of Reviews by Brand", fontsize=16)
plt.xlabel("Brand", fontsize=14)
plt.ylabel("Average Reviews", fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=12)  
plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(20, 10))
sns.barplot(x='brand', y='number_of_reviews', data=df, estimator=sum, ci=None, palette='muted')
plt.title('Total Number of Reviews by Brand')
plt.xlabel('Brand')
plt.xticks(rotation=45, ha='right', fontsize=12) 
plt.ylabel('Total Reviews')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['discount'], bins=10, kde=True, color='skyblue')
plt.title('Distribution of Discounts')
plt.xlabel('Discount (%)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='size_inch', y='discounted_price', hue='brand', style='brand', data=df, palette='deep')
plt.title('Price vs Screen Size by Brand')
plt.xlabel('Screen Size (inch)')
plt.ylabel('Discounted Price (INR)')
plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(12, 8))  # Slightly larger for better spacing

# Create scatter plot
sns.scatterplot(
    data=df,
    x='size_inch',
    y='discounted_price',
    hue='brand',
    style='brand',
    palette='deep',
    s=100,  # Increase marker size for visibility
    edgecolor='black'  # Add border to markers
)

# Add titles and labels
plt.title('Discounted Price vs. Screen Size by Brand', fontsize=16)
plt.xlabel('Screen Size (inches)', fontsize=14)
plt.ylabel('Discounted Price (INR)', fontsize=14)

# Improve layout and readability
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(title='Brand', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=10)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20, 12))
sns.boxplot(x='brand', y='discounted_price', data=df, palette='pastel')
plt.title('Price Distribution by Brand')
plt.xlabel('Brand')
plt.ylabel('Discounted Price (INR)')
plt.xticks(rotation=45, ha='right', fontsize=12) 
plt.tight_layout()
plt.show()
```

```python


# Count products per brand
brand_counts = df['brand'].value_counts()

# Limit to top N brands for clarity (optional)
top_n = 10
brand_counts = brand_counts.head(top_n)

# Create pie chart
plt.figure(figsize=(12, 8))
plt.pie(
    brand_counts,
    labels=[label.title() for label in brand_counts.index],  # Capitalize for readability
    autopct='%1.1f%%',
    startangle=140,
    colors=plt.cm.tab20.colors,
    wedgeprops={'edgecolor': 'white'},
    textprops={'fontsize': 12}  # Increase label font size
)

# Add title
plt.title('Product Distribution by Top Brands', fontsize=16)
plt.tight_layout()
plt.show()
```
